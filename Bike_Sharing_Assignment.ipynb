* BILDING A MODEL WITH ALL FEATURES USING STATSMODELS:

```python
import statsmodels.api as sm 
X_train_lm = sm.add_constant(X_train)

lr1 = sm.OLS(y_train, X_train_lm).fit()
lr1.params
```

```python
# Importing RFE and LinearRegression 
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of variable equal to 15 
lm = LinearRegression()
lm.fit(X_train, y_train)
rfe = RFE(estimator=lm, n_features_to_select=15)
rfe = rfe.fit(X_train, y_train)
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
# RFE support varibles
col = X_train.columns[rfe.support_]
print(len(col))
col
```

```python
# RFE not supported variables 
X_train.columns[~rfe.support_]
```

```python
#Let's consider only the RFE supported columns for model building
X_train_rfe = X_train[col]
```

```python
### MODEL 2
```

```python
def calculateVIF(df): 
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range (df.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = 'VIF', ascending = False)
    return vif
```

```python
calculateVIF(X_train_rfe)
```

```python
# Model 2
# Add a constant

X_train_lm2 = sm.add_constant(X_train_rfe)

# Create a first fitted model 
lr2 = sm.OLS(y_train, X_train_lm2).fit()
lr2.summary()
```

```python
# All of the variables in the abvoe analysis have low p-values, and the adjusted R-square value is 0.847, 
#up from 0.843 to 0.847 after 14 features were removed. This indicates that the majority of the features were insignificant and a small number were negative. collrelatin #from the VIP table above, 
# we can observe that the workingday has a quite high vip value of 31.53. Let's take this feature out.

```

```python
X_train_rfe2 = X_train_rfe.drop(['workingday'], axis=1)
X_train_rfe2.head()
```

```python
calculateVIF(X_train_rfe2)
```

```python
# Model 3 
```

```python
# Add a constant 
X_train_lm3 = sm.add_constant(X_train_rfe2)

#Create the first fitted model 
lr3 = sm.OLS(y_train, X_train_lm3).fit()
lr3.summary()
```

```python
# As we can see from the foregoing, weekday_sat has a high pValue of 0.482>0.05 and a hum VIP of 14.57>5.
#Leave weekday_sat.
```

```python
X_train_rfe3 = X_train_rfe2.drop(['weekday_sat'], axis = 1)
X_train_rfe3.head()
```

```python
calculateVIF(X_train_rfe3)
```

```python
# Model 4 
X_train_lm4 = sm.add_constant(X_train_rfe3)

#Create the first fitted model 
lr4 = sm.OLS(y_train, X_train_lm4).fit()
lr4.summary()
```

```python
#for each feature, we can observe that the p-value is less than 0.05 and the adj. R-squared is 0.847.
# VIP for hum is 14.57>5, so let's eliminate the hum first.

```

```python
X_train_rfe4 = X_train_rfe3.drop(['hum'], axis = 1)
X_train_rfe4.head()
```

```python
calculateVIF(X_train_rfe4)
```

```python
X_train_lm5 = sm.add_constant(X_train_rfe4)

lr5 = sm.OLS(y_train, X_train_lm5).fit()
lr5.summary()
```

```python
# It is evident from the Model Summary Report above that each feature's p-value is less than 0.05.
# Note: This is a heuristic technique that suggests examining a variable from a business standpoint, analyzing any outliers, and dropping a feature if the VIF value of any predictor is between 5 and 10. If the VIF value is larger than 10.
# Now that we can solve this 
```

### LINEAR REGRESSION ASSUMPTION PROOFS 

#### Residual analysis

```python
# Perform residual analysis
y_train_pred = lr5.predict(X_train_lm5)
residual = y_train_pred - y_train
```

```python
# Assumption of normally distributed error term

# Plot histogram 
sns.distplot(residual)
plt.xlabel('Residual (Error Term)', fontdict={'fontsize': 20})
plt.title('Error Distribution', fontdict={'fontsize': 25})
plt.show() 
```

```python
# From the above graph it is evident that Error Distribution Is Normallly Distributed Across 0, 
# which indicates that our model has handled the assumption of Error Normal Distribution properly.
```

```python
# Assumption of Error Terms Being Independent

sns.regplot(x=y_train_pred, y=residual)
plt.title('Residual Vs. Predicted Values (Pattern Identification)', fontdict={'fontsize': 20})
plt.xlabel('Predicted Values', fontdict={'fontsize': 15})
plt.ylabel('Residuals', fontdict={'fontsize': 15})
plt.show()
```

```python
# no relation between Residual & Predicted Value
```

```python
# Homoscedasticity

sns.regplot(x=y_train, y=y_train_pred)
plt.title('Predicted Points Vs. Actual Points', fontdict={'fontsize': 20})
plt.xlabel('Actual Points', fontdict={'fontsize': 15})
plt.ylabel('Predicted Points', fontdict={'fontsize': 15})
plt.show()
```
